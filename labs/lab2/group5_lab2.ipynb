## Our definition of "outliers"

TowardsDataScience describes outliers as “extreme values that deviate from other observations on data [which] may indicate a variability in a measurement, experimental errors or a novelty. In other words, an outlier is an observation that diverges from an overall pattern on a sample” (https://towardsdatascience.com/a-brief-overview-of-outlier-detection-techniques-1e0b2c19e56). Outliers can be caused by human, measurement, experimental, data processing, sampling, and natural error. If someone received these paychecks -- $225, $250, $25, $235 -- $25 would be an outlier value. Some measures are skewed by outliers like the mean and standard deviation. As a result, removing outliers can lead to more accurate analysis of a particular dataset. 

## Peirce's Criterion

Peirce’s Criterion, developed by Benjamin Peirce in 1852, is a derivation method based on probability theory that can be used to eliminate data “outliers” (see above) in a rational manner. It was based on the following principle: “the proposed observations should be rejected when the probability of the system of errors obtained by retaining them is less than that of the system of errors obtained by their rejection multiplied by the probability of making...abnormal observations” (Peirce). The actual method of calculation used by Peirce was mathematically tedious to use and B.A. Gould later took Peirce’s Criterion and presented it in a more usable format with tables he derived from Peirce’s work. 

Peirce’s method to determine whether data values should be rejected involves the procedure as follows:
1. Calculate the mean and the sample standard deviation of the complete data set
Obtain R corresponding to the number of measurements taken from Peirce’s table
2. Assume the case of one doubtful observation first, even if there is seemingly be more than one $ R = |x_i-x_m|$max $÷ \sigma $
3. R represents the ratio of the maximum allowable deviation of a measured value from the data mean to the standard deviation
4. $ \sigma $ is the “sample” standard deviation of the complete data set
$ x_i $ is a measured data value. $ x_m $ is the mean of the data set
5. Calculate the maximum allowable derivation: $ |x_i-x_m| $ max 
6. For any suspicious data measurements, obtain $ |x_i-x_m| $ 
7. Eliminate the suspicious measurements if: $ |x_i-x_m| > |x_i-x_m| $ max 
8. If this results in the rejection of one measurement, assume the case of two doubtful observations, keeping the original values of the mean and standard deviation, and the original number of measurements. In other words, go to step 8
If more than one measurement is rejected in the above test, assume the value of doubtful observations. For example, if two measurements are rejected in step 5, assume the case of three doubtful observations, keeping the original values of the mean, standard deviation, and the original number of measurements as the process is continued
9. Repeat steps 2-5, sequentially increasing the number of doubtful measurement possibilities, until no more data measurements need to be eliminated
Obtain the new value of the mean and sample standard deviation of the reduced data set 

## One other method for outlier identification: Chauvenet’s Criterion

Information below on Chauvenet’s criterion comes from the Chauvenet’s criterion Wikipedia article and the Statistics How to entry on Chauvenet’s criterion. Information coming from other sources is cited specifically.
	Chauvenet’s Criterion is another method for identifying outliers that is similar to but not the same as Peirce’s Criterion. Like Chauvenet’s criterion, determining whether a value is an outlier does not depend on specific values of the data, “only characteristics of the observation data”; it is based on the mean and sample size of the data (Peirce). The idea behind Chauvenet’s criterion is that it determines an acceptable range of values, known as a probability band, based on the mean of the dataset. Values that lie outside this range are considered outliers. The formula for calculating the probability comes from the idea that the band must include n-½ samples. The formula for the probability equals (n-½)/n or, simplified, 1 - 1/2n. The value 1/2n is equivalent to the probability of the two tails of the normal distribution. To calculate Chauvenet’s criterion, calculate the mean and standard deviation, find the difference between the potential outlier and the mean, and calculate the probability of a data point that many standard deviations away occurring. Multiple this probability by n. If it is less than .5, the data point is an outlier. Chauvenet’s criterion was developed a few years after Peirce’s criterion. However, Peirce’s criterion is still more rigorous. Chauvenet’s criterion is also based on the “arbitrary assumption that a measurement may be rejected if the probability of obtaining the deviation from the mean for that value is less than the inverse of twice the number of measurements.” While Peirce’s criterion can be run to determine multiple outliers, Chauvenet’s criterion is generally used to determine if a single point is an outlier. That is, Chauvenet’s criterion does not have a method of distinction for whether there are multiple suspected outliers. However, Chauvenet’s criterion is also easier to calculate, especially in the case of merely checking if a single point is an outlier. 

## About FLUXNET and its 2015 dataset...

FLUXNET is a collective network of towers that measure energy and gas fluctuations between the Earth's surface and the atmosphere. The measurements made by this network of towers provide useful information that can help our models predict and detect trends that may occur in weather, climate, carbon and water cycles, and even air pollution. 

The FLUXNET 2015 Dataset includes data collected at sites from multiple regional flux networks. The FLUXNET 2015 Dataset is an improved dataset from the FLUXNET Marconi Dataset (2000) and the FLUXNET LaThuile Dataset (2007), with better “data quality control protocols and the data processing pipeline.” These improvements include close interactions within tower teams, new methods for uncertainty quantification, and use of reanalysis data to fill long gaps of micrometeorological variable records, and other improvements. 

The FLUXNET2015 dataset provides ecosystem-scale data on CO2, water, and energy exchange between the biosphere and the atmosphere, and other meteorological and biological measurements, from 212 sites around the globe (over 1500 site-years, up to and including year 2014). These sites, independently managed and operated, voluntarily contributed their data to create global datasets. Data were quality controlled and processed using uniform methods, to improve consistency and intercomparability across sites. The dataset is already being used in a number of applications, including ecophysiology studies, remote sensing studies, and development of ecosystem and Earth system models. FLUXNET2015 includes derived-data products, such as gap-filled time series, ecosystem respiration and photosynthetic uptake estimates, estimation of uncertainties, and metadata about the measurements (Pastorello). 

# Methods

The methodology to remove outliers using Peirce’s criterion combined “outlier.py” (Python code) from the GePiSaT documentation and “peirces_dev” (R code) from the Peirce's criterion Wikipedia page. 

The Python script (“outlier.py”) from the GePiSat documentation performs an initial check for outliers, a secondary check for outliers in the case where none are initially found, and an iterative search for additional outliers by incrementing Peirce’s n parameter until the number of outliers identified is less than the number of outliers assumed. Finally, the script removes the outliers from the original datasets. The R script (“peirces_dev”) from the R implementation section of the Wikipedia page uses Peirce’s criterion to return the squared maximum error deviation, x2. Then, x2 must be multiplied by the mean-squared error of the model fit to get the squared-delta value, Δ2. Any observation pairs with a squared-error value greater than Δ2 are considered outliers and can be removed.  

After reproducing the python script (“outlier.py”) from the GePiSat documentation in R, it can be used in conjunction with the R script (“peirces_dev”) to identify and highlight outliers from the dataset in the plot. 
 



```
# impport necessary library
install.packages('latex2exp')
library(latex2exp)
```

```
# read in data 
flux.url <- "https://ds-wm.github.io/course/atsa/data/FLX_US-Goo_FLUXNET2015_SUBSET_HH_2002-05_1-4.csv"
df <- read.csv(flux.url)
df$index <- c(1:length(df$SW_IN_F))
# create a linear model where SW_IN_F is dependent on NEE_VUT_REF and find residuals
lin.mod <- lm(df$SW_IN_F ~ df$NEE_VUT_REF)
residuals <- lin.mod$residuals
```

```
plot(df$NEE_VUT_REF, df$SW_IN_F, xlab = 'NEE', ylab=TeX("$ SW_{IN} $"))
```

```
# below is the peirce_dev function taken directly from the Peirce's criterion Wikipedia page
peirce_dev <- function(N, n, m) {
    # N :: total number of observations
    # n :: number of outliers to be removed
    # m :: number of model unknowns (e.g., regression parameters)
    #
    # Check number of observations:
    if (N > 1) {
       # Calculate Q (Nth root of Gould's equation B):
       Q = (n^(n/N) * (N-n)^((N-n)/N))/N
       #
       # Initialize R values:
       Rnew = 1.0
       Rold = 0.0  # <- Necessary to prompt while loop
       #
       while (abs(Rnew-Rold) > (N*2.0e-16)) {
           # Calculate Lamda (1/(N-n)th root of Gould's equation A'):
           ldiv = Rnew^n
           if (ldiv == 0) {
              ldiv = 1.0e-6
           }
           Lamda = ((Q^N)/(ldiv))^(1.0/(N-n))
           #
           # Calculate x-squared (Gould's equation C):
           x2 = 1.0 + (N-m-n)/n * (1.0-Lamda^2.0)
           #
           # If x2 goes negative, set equal to zero:
           if (x2 < 0) {
              x2 = 0
              Rold = Rnew
           } else {
              #
              # Use x-squared to update R (Gould's equation D):
              # NOTE: error function (erfc) is replaced with pnorm (Rbasic):
              # source: 
              # http://stat.ethz.ch/R-manual/R-patched/library/stats/html/Normal.html
              Rold = Rnew
              Rnew = exp((x2-1)/2.0)*(2*pnorm(sqrt(x2)/sqrt(2)*sqrt(2), lower=FALSE))
           }
       }
    } else {
       x2 = 0
    }
    x2
}
```

```
# below is our modified version of the outliers script in the GePiSaT documentation, converted to R
n <- length(lin.mod$residuals) # Number of observations
se <- (lin.mod$residuals)^2 # Squared error 
mse <- (sum(se))/(n-2) # Mean Squared Error using GEPISAT Formula
x2 <- peirce_dev(n, 1, 2) 
d2 <- mse * x2 
n_index <- se[se>d2]
n_found <- length(n_index)
outlier_index <- which(se>d2)
n_found <- length(outlier_index)
if (n_found==0) {
  x2 = peirce_dev(n, 2, 2)
  d2 = mse*x2
  outlier_index <- which(se>d2) 
  n_index <- which(se>d2)[1]
  n_found <- length(outlier_index)
}
j <- 1
while(j <= n_found) {
  j <- j+1
  x2 <- peirce_dev(n, j, 2) 
  d2 <- mse*x2
  outlier_index <- which(se>d2) 
  n_index <- which(se>d2)[1]
  n_found <- length(outlier_index)
}
```

# Results

Below is a plot showing the linear relationship between net ecosystem exchange ($NEE$) and incoming shortwave radiation ($SW_in$). The figure below depicts the outliers identified by applying Peirce's criterion in red. 

```
outliers_nee <- subset(df,(df$index %in% outlier_index))$NEE_VUT_REF 
outliers_sw <- subset(df,(df$index %in% outlier_index))$SW_IN_F
plot(df$NEE_VUT_REF, df$SW_IN_F, xlab = 'NEE', ylab=TeX("$ SW_{IN} $"), main = "Outliers Identified Using Peirce's Criterion on FLUX Dataset")
abline(lin.mod)
points(outliers_nee, outliers_sw, col="red")
```
